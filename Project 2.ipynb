# Analysis of Pythagorean Triples

## Introduction

_Pythagorean triples_ are a set of three positive numbers $a, b, c$ such that

$$a^2 + b^2 = c^2$$

where $a$, $b$, and $c$ all form the sides of a right triangle.

A great way to picture this is to connect three squares, $A$, $B$, and $C$ by the corners to form an area in the middle of them shaped like a triangle. This is where the formula $a^2 + b^2 = c^2$ comes from. Each square has an area of $L \times W$ where $L = W$. The sum of areas of the two smaller squares $A$ and $B$ is equal in area to the largest square $C$.

Some examples of Pythagorean triples are:

* $(3, 4, 5)$

* $(4, 3, 5)$

* $(5, 12, 13)$

If $\sqrt{a^2 + b^2}$ is equal to an integer $c$, then the integers $a$ and $b$ are called *Pythagorean tuples*. The goal of this project is to analyze patterns and trends of Pythagorean tuples.

## Finding Pythagorean Triples

In this section, we will create a function called `ptriples(n)` that will return a list of Pythagorean triples.

```python
from math import sqrt                                   # import the square root function from the math library
from matplotlib import pyplot as plt                    # import pyplot from matplotlib
```

```python
def ptriples(n):                                        # Define the function ptriples
    a = range(1, n+1)                                   # Create a variable "a" with range [1, n+1)
    b = range(1, n+1)                                   # Create a variable "b" with range [1, n+1)
    sol = []                                            # Create an empty list called sol for the Pythagorean triples
    for x in a:                                         # loop over the "a" range
        for y in b:                                     # loop over the "b" range
            if sqrt(x**2+y**2)-int(sqrt(x**2+y**2))==0: # if the the differnce of squares = 0
                                                          # Note: The int function rounds floats down         
                c = [x, y, int(sqrt(x**2 + y**2))]      # Create list for values in the Pythagorean triple
                sol.append(c)                           # Add that triple to our sol list
    return sol                                          # Return sol list

ptriples(50)
```

## Plotting Pythagorean triples

A function called `plotter(n)` is used to grab the Pythagorean triples from the previous section, take their respective Pythagorean tuples $(a, b)$, and plot them on two scatter plots. One plot is graphed with $a$ on the x-axis and $b$ on the y-axis. The other plot has $a$ on the y-axis and $b$ on the x-axis. Each plot has the line $y = x$ plotted on them as well.

```python
def plotter(n):                                        # Define a plotting function
    t = ptriples(n)                                    # Create a variable "t" that is a list of triples
   
    a = []                                             # Create an empty list for the a values in the tuple
    b = []                                             # Create an empty list for the b values in the tuple
    r = range(0, n)                                    # Create a range variable called "r" ranging from [0, n)
    for s in t:                                        # For the list values in t
        a.append(s[0])                                 # Append the a values of the triples into the a list
        b.append(s[1])                                 # Append the b values of the triples into the b list
        
    plt.figure(figsize=(15,5))                         # We set the plot size of each graph to 15 inches by 5 inches
    plt.subplot(1,2,1)                                 # Initiate a subplot in the first section of a 1x2 grid
    plt.xlabel('$a$', fontsize=15)                     # Create an "a" label with a font size of 15
    plt.ylabel('$b$', fontsize=15)                     # Create a "b" label with a font size of 15
    plt.title("Pythagorean Tuples (a,b)", fontsize=20) # Create a title for the graph with a fontsize of 20
    plt.plot(a, b, "b.")                               # Plot "a" and "b" using blue dots
    plt.plot(r, "r-")                                  # Create a red line for y = x 
    
    plt.subplot(1,2,2)                                 # Initiate a subplot in the second section of a 1x2 grid
    plt.xlabel('$b$', fontsize=15)                     # Create an "a" label with a font size of 15
    plt.ylabel('$a$', fontsize=15)                     # Create a "b" label with a font size of 15
    plt.title("Pythagorean Tuples (b,a)", fontsize=20) # Create a title for the graph with a font size of 20
    plt.plot(b, a, "b.")                               # Plot "a" and "b" using blue dots
    plt.plot(r, "r-")                                  # Create a red line for y = x
    
plotter(500)    
```

## Trends and Structures of Pythagorean Tuples

There are couple of trends that can be discussed. The first thing I notice is that the Pythagorean tuples are placed in such a way that they are reflected over the line $y = x$. This is why I invert the x and y axis. Another thing I noticed is that as you plot more Pythagorean tuples, you can see that more lines are formed in the data. Finally, you can see that the tuple points create a pattern that look like rings that surround $y = x$. The rings get larger and larger as you continue along $y = x$.
